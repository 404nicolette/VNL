```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv("vnl.csv")
original_data
```

```python
print(original_data.isnull().sum())
```

```python
stat_cols = ["attack","block","serve","set","dig","receive"]
per_country = original_data.groupby("country")[stat_cols].mean()
per_country
```

```python
def get_countries(original_data):
    country_list = []
    for i in original_data["country"]:
        if i not in country_list:
            country_list.append(i)

    return(country_list)
countries_list = get_countries(original_data)
print(countries_list)
```

```python
def get_stat_columns(original_data):
    stat_cols = original_data.select_dtypes(include='number').columns
    
    stat_list = []
    
    for col in stat_cols:
        if col not in stat_list:
            stat_list.append(col)
    del stat_list[0]
    
    return stat_list

stat_list = get_stat_columns(original_data)
print(stat_list)
```

```python
# mean performances
stat_list = get_stat_columns(original_data)

for stat in stat_list:
    
    plt.figure(figsize=(6, 3))

    sns.barplot(x=per_country.index, y=stat, data=per_country)
    plt.xticks(rotation=90)
    plt.title(f'Average {stat.capitalize()} Performance by Country')
    plt.show()
```

```python
# heat map for all perfomance
plt.figure(figsize=(10, 8))

heatmap = sns.heatmap(per_country, annot=True, cmap='coolwarm', linewidths=0.5, cbar=True)

# label
cbar = heatmap.collections[0].colorbar
cbar.set_label('Performance Score')

plt.title('Country Performance Heatmap')
plt.xlabel('Performance Metrics')
plt.ylabel('Country')

plt.show()
```

```python
'''
based on statistical aveage alone, Italy is not dominating in every department-- leading only in service in constrast
to Serbia who is leading in multiple department. Serbia who finished 11th in the overall tournament does not reflect
this domination on the rankings. 

Italy leading the service department and eventually winning the gold further solidifies my personal beliefs that 
service wins games. When a team generally perfoms well on service it creates a domino effect where the opponent's defense
pattern is not as solid as it would be if the team has a weak service game. Good service has two outcomes: an ace or
a bad reception. Bad reception tends to give low-quality sets which affects opponent's attack. However, I would like to
acknowledge that this is not always the case, there are exceptional setters or attackers or both who are still able
to convert bad reception into points (ie Vargas).
'''
```

```python
countries_list = get_countries(original_data)
stat_list = get_stat_columns(original_data)

for stat in stat_list:

    plt.figure(figsize=(6, 3))
    sns.boxplot(x='country', y=stat, data=original_data)
    plt.xticks(rotation=90)
    plt.title(f'Average {stat.capitalize()} Performance between Countries')
    plt.show()
```

```python
# extremes per country

stat_list = get_stat_columns(original_data)
numeric_cols = original_data.select_dtypes(include='number').columns

country_performance_stats = original_data.groupby('country')[numeric_cols].agg(['median', 'std', 'min', 'max'])


for stat in stat_list:
    plt.figure(figsize=(12, 8))

    heatmap_data = country_performance_stats[[(stat, 'min'), (stat, 'max'), (stat, 'std')]]


    sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt=".2f")
    plt.title(f"Min, Max, and Std Dev of {stat.capitalize()} Performance by Country")
    plt.xlabel('Statistic')
    plt.ylabel('Country')
    plt.xticks(rotation=45)
    plt.show()
```

```python
'''
based on the extremes, it is clear that the min calculations per department is skewed as as players who plays different
positions such as Liberos who doesnt attack are still included in the calculation causing the results to not be as
reflective.

in conclusion, as much as sports analytics is evolving-- there's are factors in sports that numbers and data cannot
capture. the tenancity, the team chemistry, the momentum that occurs each games.
'''
```
